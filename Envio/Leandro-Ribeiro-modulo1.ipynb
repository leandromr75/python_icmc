Nesses exercícios estaremos lidando com dados reais de gastos de deputados. Temos informações referentes a diversos pedidos de reembolso, tendo o nome do deputado, seu estado, seu partido e o valor do reembolso solicitado.

Ignore o código a seguir por enquanto, estamos somente carregando os dados. Logo veremos tudo que é necessário para entender como ele funciona.

````python
def carregar_dados():
    """
    Função que carrega a base de dados como uma lista de tuplas
    """
    with open('gastos_deputados.csv', 'r',encoding = 'utf-8') as f:
        content = f.readlines()
    content = [x.strip() for x in content]
    data = [line.split(',') for line in content]
    for line in data:
        line[-1] = float(line[-1])
    data = [tuple(line) for line in data]

    return data
````

O código abaixo carrega os dados que iremos trabalhar. Eles estão organizados em uma lista de tuplas, onde cada tupla contém as informações de um reembolso. Como a seguir:

```python
dados = [(NOME, ESTADO, PARTIDO, VALOR),
         (NOME, ESTADO, PARTIDO, VALOR),
         ...
         (NOME, ESTADO, PARTIDO, VALOR)]
```

````python
dados = carregar_dados()
````

````python
# questões 1 e 2
from statistics import mean
soma = 0
valor_medio = []
# calcular valor total e médio solicitado em reembolsos
for valor in dados:
    soma += valor[3]
    valor_medio.append(valor[3])

# converter para valor monetário
valor_m = mean(valor_medio)

import locale
locale.setlocale(locale.LC_MONETARY, 'pt_BR.UTF-8')
soma = locale.currency(soma , grouping=True)
valor_m = locale.currency(valor_m , grouping=True)
print('Valor total dos reembolsos: ' + soma)
print('Valor médio dos reembolsos: ' + valor_m)

````

````python
# questão 3
GO = 0
MT = 0
for estado in dados:
    if estado[1] == 'GO':
        GO += estado[3]
    if estado[1] == 'MT':
        MT += estado[3]

import locale
locale.setlocale(locale.LC_MONETARY, 'pt_BR.UTF-8')
GO = locale.currency(GO , grouping=True)
MT = locale.currency(MT , grouping=True)

if GO > MT:
    print('Goiás tem um valor maior em reembolsos: ' + str(GO))
elif MT > GO:
    print('Mato Grosso tem um valor maior em reembolsos: ' + str(MT))
else:
    print('Empatados os estados de GO e MT no valor de reembolsos: ' + str(GO))
````

````python
# questão 4
from statistics import mean
Sudeste = []
Sul = []
for estado in dados:
    if estado[1] == 'RS':
        Sul.append(estado[3]) 
    elif estado[1] == 'SC':
        Sul.append(estado[3])
    elif estado[1] == 'PR':
        Sul.append(estado[3])
    elif estado[1] == 'SP':
        Sudeste.append(estado[3])
    elif estado[1] == 'RJ':
        Sudeste.append(estado[3])
    elif estado[1] == 'MG':
        Sudeste.append(estado[3])
    elif estado[1] == 'ES':
        Sudeste.append(estado[3])


import locale
locale.setlocale(locale.LC_MONETARY, 'pt_BR.UTF-8')

if mean(Sul) < mean(Sudeste):
    print('Região Sul tem um valor médio menor em reembolsos: ' + str(locale.currency(mean(Sul) , grouping=True)))
elif mean(Sudeste) < mean(Sul):
    print('Região Sudeste tem um valor menor em reembolsos: ' + str(locale.currency(mean(Sul) , grouping=True)))
else:
    print('Regiões empatadas no valor de reembolsos: ' + str(locale.currency(mean(Sul) , grouping=True)))
````
