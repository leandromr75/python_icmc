No de Pandas, houveram algumas dúvidas sobre como ler o dataframe:

```python
# Vamos ler os dados
import pandas as pd
df = pd.read_csv("gdp_per_capita_ppp.csv")
df.head()
```

```python
# Tirando a 1a linha, a transposta já quase nos fornece o que queremos:
new_df = df.T
new_df.head()
```

```python
# Podemos então usar a primeira linha como o nome das colunas:
new_df.columns = new_df.iloc[0]
# Retirar a primeira linha
new_df = new_df.iloc[1:]
# Configurar o index
new_df.index = df.columns[1:]
```

```python
new_df.tail()
```

```python
# Agora calcular a média:
new_df.fillna(0).mean(axis=0).to_frame()
```
